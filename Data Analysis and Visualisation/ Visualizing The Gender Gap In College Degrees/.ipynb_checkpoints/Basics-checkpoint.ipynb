<h1>Studying the gender gap in college degrees</h1>

In this notebook I will be demonstrating how we can use data visualisation to show the proportion of each gender chosing to study a range of college degrees. In the last step of this project, I'll explore how to export the final diagram we create as an image file.

I will first show the difference in gender for STEM degrees

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

I have ordered the charts above in such a way to reduce the Data-Ink ratio and make the difference in gender for each degree category as clear as possible. As you can see, engineering has the greatest 'gender gap'. All of the STEM categories have demonstrated a declining gender gap except for Computer Science and Psychology. The gender gap in Computer science decreased in the 90s but has increased in recent years. The gender gap for psychology has increased over time with a decrease in the number of men chosing to study this disciplne, but an increase in the number of women.

```python
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16,16))
#fig.subplots_adjust(hspace=.5)

for sp in range(0, 18, 3):
    category_idx = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
        
for sp in range(1, 16, 3):
    category_idx = int((sp-1)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
for sp in range(2, 20, 3):
    category_idx = int((sp-2)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
```

The graphs above give us a neat comparison of the difference in the gender of students across all degree subjects. The graphs have been displayed in such a way that subjects with the smallest gender gap are displayed at the bottom, and those with the largest gender gap are displayed at the top. This allows your eye to browse from top to bottom and gain a quick understanding of the data.

With seventeen line charts in one diagram, the non-data elements quickly clutter the field of view. The most immediate issue that sticks out is the titles of some line charts overlapping with the x-axis labels for the line chart above it. If we remove the titles for each line chart, the viewer won't know what degree each line chart refers to. Let's instead remove the x-axis labels for every line chart in a column except for the bottom most one. We can accomplish this by modifying the call to:

`Axes.tick_params(labelbottom='off')`

Alternatively we could use the subplot_adjust method to offset the graphs, moving each one down so they do not overlap:

`fig.subplots_adjust(hspace=.5)`

In addition to the x-axis overlay, the y-axis labels look a bit cluttered, to solve this issue, we can use the following command to remove the y-tick values between 0 and 100 and just keep the minumum and maximum value:

`ax.set_yticks([0,100])`

It would also be useful to have a good visual indicator of which degrees have a gender breakdown close to 50-50. To do this I will add a line on the y-axis label 50. I will use <a href='http://tableaufriction.blogspot.co.uk/2012/11/finally-you-can-use-tableau-data-colors.html'>the Color Blind 10 palette</a> to chose an appropriate color, and make it slightly transparent.

```python
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(16,16))
#I could use the following to offset the graphs instead of removing
#the x-axis
#fig.subplots_adjust(hspace=.5)

for sp in range(0, 18, 3):
    category_idx = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #Here I am setting the graph to display only the min and max
    #values on the y-axis
    ax.set_yticks([0,100])
    #Adding a line at the 50 point
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    ax.set_title(stem_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.tick_params(labelbottom='off')
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        #I have turned the x-axis off except for the last graph
        #of each group
        ax.tick_params(labelbottom='on')
        
        
for sp in range(1, 16, 3):
    category_idx = int((sp-1)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    ax.set_title(lib_arts_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.tick_params(labelbottom='off')
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 4:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
for sp in range(2, 20, 3):
    category_idx = int((sp-2)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[category_idx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[category_idx]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    ax.set_title(other_cats[category_idx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.tick_params(labelbottom='off')
    if category_idx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif category_idx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
#Now I will save thge plot to a png file
plt.savefig("gender_degrees.png")
```
